```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
 
# 创建生成器对象
generator = countdown(5)
 
# 通过迭代生成器获取值
print(next(generator))  # 输出: 5
print(next(generator))  # 输出: 4
print(next(generator))  # 输出: 3
 
# 使用 for 循环迭代生成器
for value in generator:
    print(value)  # 输出: 2 1
```

```python
import numpy as np

# 用户偏好矩阵（3用户 × 3维度）
U = np.array([
    [0.9, 0.1, 0.8],
    [0.2, 0.9, 0.3],
    [0.5, 0.5, 0.5]
])

# 电影特征矩阵（4电影 × 3维度）
M = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 1],
    [0, 0, 1]
])

# 点积：用户 × 电影.T = 每个用户对每个电影的打分
# scores = np.dot(U, M.T)
scores = U @ M.T

print(scores)
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# 矩阵乘法
C = np.dot(A, B)
print("矩阵乘积：\n", C)

# 转置
print("转置矩阵：\n", A.T)

# 逆矩阵
inv_A = np.linalg.inv(A)
print("逆矩阵：\n", inv_A)

# 行列式
det_A = np.linalg.det(A)
print("行列式：", det_A)
```

```python
# 创建一组随机成绩数据（10人，3门课）
import numpy as np
import pandas as pd

df = pd.DataFrame({
    '姓名': [f'同学{i}' for i in range(1, 11)],
    '数学': np.random.randint(60, 100, size=10),
    '英语': np.random.randint(60, 100, size=10),
    '语文': np.random.randint(60, 100, size=10)
})

# 计算平均分
df['平均分'] = df[['数学', '英语', '语文']].mean(axis=1)

# 按平均分排序
df = df.sort_values(by='平均分', ascending=False)

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc("font", family='Microsoft YaHei')

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contour(X, Y, Z, levels=20)
plt.colorbar()
plt.title("等高线图：Z = X² + Y²")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()
```
